Using matplotlib to make pretty graphs
-

```python
from numpy import *
```

```python
t = arange(0.0, 10.0, .1)
```

```python
x = sin(2 *pi * t)
s = cos(2* pi *t)
```

```python
plot(t, x, linewidth= 2, label ='$sin(4 \pi t)$')
plot(t, s, color= 'red', label = '$cos(4 \pi t)$')
xlabel('Time (s)', fontsize= 15)
ylabel('Voltage (mV)', fontsize = 15)
title('Voltage vs Time', fontsize=20)
legend()
```
