```python
def eratosthenes_sieve(maximum):
    is_prime = ones((maximum,), dtype=bool)
    is_prime[0] = 0
    is_prime[1] = 0
    N_max = int(sqrt(len(is_prime)))
    for j in range(2, N_max):
        is_prime[2*j::j] = False
    return is_prime.nonzero()
```

```python
eratosthenes_sieve(100)
```

```python
eratosthenes_sieve(1000)
```
