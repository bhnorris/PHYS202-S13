Electric Dipole Potential 
-
$
V(x,y) = \dfrac{kq}{\sqrt{x^2 + (y - d)^2}} - \dfrac{kq}{\sqrt{x^2 + (y + d)^2}}
$

```python
k, q, d = 1,1,1
```

```python
dx, dy = .06, .06
x = arange(-2, 2, dx)
y = arange(-2, 2, dy)
print x.ndim
print len(x)

X,Y = meshgrid(x,y)
V = (k * q) / (sqrt(X**2 + (Y - d)**2)) - (k * q) / (sqrt(X**2 + (Y + d)**2))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, lw= 0, cmap= cm.Spectral)
cb = fig.colorbar(p, shrink= .5)
```

```python
levels = array([8, 4, 2, 1, .5, .25, 0 , -.25, -.5, -1, -2, -4, -8])
levels = linspace(8.03, -8.012, num = 1000)
contour(X,Y,V,levels,linewidths = 2, cmap=cm.flag)
```

EX 1 Make a 3d plot of $G(x,y) = e^{(-x^2 - y^2)}$ over the range -6 < $(x, y)$ < 6 using 51 bins in $x$ and 51 bins in $y$.

EX 2 Determine the electric potential for a quadrupole consisting of q = -1.0 nC at both (+-2, 0) and q = +1.0 nC at both (0, +-2) Make a surface plot and a contour plot on the range -4 < (x, y) < 4 using 101 bins in the x and 101 bins in the y what is the value of the function at V(20, 30)

```python
k, q, d = 8.9875517*10**9,(1*10**(-9)),2
#k in N m^2 C^-2 q in C and d in m
dx, dy = (8/101.0), (8/101.0)
x = arange(-4, 4, dx)
y = arange(-4, 4, dy)


X,Y = meshgrid(x,y)
V = (k * q) / (sqrt(X**2 + (Y - d)**2)) + (k * q) / (sqrt(X**2 + (Y + d)**2)) - (k * q) / (sqrt(Y**2 + (X - d)**2)) - (k * q) / (sqrt(Y**2 + (X + d)**2))
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, lw= 0, cmap= cm.jet)
cb = fig.colorbar(p, shrink= .5)
```

```python
levels = array([8, 4, 2, 1, .5, .25, 0 , -.25, -.5, -1, -2, -4, -8])

contour(X,Y,V,levels,linewidths = 2, cmap=cm.jet)
```
