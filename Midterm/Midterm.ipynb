```python
def is_anagram(s1, s2):
    x = []
    y = []
    if len(s1) != len(s2):
        return False
    for letter in s1:
        x.append(letter.lower())
    for letter in s2:
        y.append(letter.lower())
    x.sort()
    y.sort()
    if x == y:
        return True
    else:
        return False
```

```python
print is_anagram('stan', 'nats')
```

```python
print is_anagram('murder', 'redrum')
```
