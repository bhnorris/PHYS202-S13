```python
f = lambda x: lmbd * exp(- lmbd * x)
lmbd = 1

x = arange(0, 5, .01)
a = f(x)
b = cumsum(a)/ sum(a)
N = 10000
z = numpy.random.random_sample(N);
xi = numpy.array([])
for zi in z:
    xi = numpy.append(xi,x[numpy.searchsorted(b, zi)])
hist(xi,50,normed =True)
plot(x,f(x),lw=3,color='red')
xlim(0,9.)
ylim(0,1.)
show()
```

```python
def distribute1D(x,prob,N):
    """takes any numerical distribution prob, on the interval defined by the array x,
    and returns an array of N sampled values that are statistically
    the same as the input data. """
    a = prob(x)
    b = cumsum(a)/ sum(a)
    N = 10000
    z = numpy.random.random_sample(N);
    xsamples = numpy.array([])
    for zi in z:
        xsamples = numpy.append(xsamples,x[numpy.searchsorted(b, zi)])
    
    return xsamples
```

```python
f = lambda x: lmbd * exp(- lmbd * x)
lmbd = 1

x = arange(0, 5, .01)
N = 10000

xi = distribute1D(x,f,N)
hist(xi,50,normed =True)
plot(x,f(x),lw=3,color='red')
xlim(0,9.)
ylim(0,1.)
show()
```
