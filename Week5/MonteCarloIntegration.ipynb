\begin{equation}
f(x) = x^{1/2} + \sin(x)
\end{equation}

```python
f = lambda x: x**(.5) + sin(x)
```

```python
x = arange(0.,15.,0.001)
```

```python
myplot = plot(x, f(x), lw=3)
ylabel(r'$x^{1/2} + \sin(x)$', fontsize = 20)
xlabel(r'$x$', fontsize = 20)
xlim(0.,15.)

show()
```

```python
xmin = 1.7
xmax = 12.4

subx = x[logical_and((x>xmin),(x<xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

myplot = plot(x,f(x),lw=3)
xlim(0.,15.)
plot([xmin,xmin], [ymin,ymax],color='k', linestyle='--')
plot([xmax,xmax],[ymin,ymax], color='k', linestyle='--')
plot([xmin,xmax],[ymax,ymax], color= 'k',linestyle='--')
```

```python
N= 10000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);

myplot = plot(x,f(x),lw=3)
xlim(0.,15.)
plot(samples_x[1:200],samples_y[1:200],'.')

plot([xmin,xmin], [ymin,ymax],color='k', linestyle='--')
plot([xmax,xmax],[ymin,ymax], color='k', linestyle='--')
plot([xmin,xmax],[ymax,ymax], color= 'k',linestyle='--')

show()
```

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

```python
myplot = plot(x,f(x),lw=3)
xlim = (0., 15.)

plot(samples_x[1:200],samples_y[1:200], '.')
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')
plot([xmin,xmin], [ymin,ymax],color='k', linestyle='--')
plot([xmax,xmax],[ymin,ymax], color='k', linestyle='--')
plot([xmin,xmax],[ymax,ymax], color= 'k',linestyle='--')
show()
```

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

```python
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = 100*(Iexact-I)/Iexact
print "I_exact =", Iexact
print "I_hitormiss =", I
print "Error on I =", Ierr, "%"
```

on average simply adding to the sample size didn't usually decrease error

\begin{equation}
I_N = (b - a)\langle f \rangle = (b - a) \dfrac{1}{N} \sum^{N}_{i=1}f(x_i)
\end{equation}

```python
from numpy import *
f = lambda x: x**(.5) + sin(x)
xmin = 1.7
xmax = 12.4
x = arange(0.,15.,0.001)
N = 10**(6)
smmI = 0
y = zeros(len(x),dtype='float')
for i in range(N):
    samples_x = xmin + (xmax-xmin)*random_sample(size=None)
    smmI += f(samples_x)
smmI = smmI/N
for n in range(len(y)):
    y[n] = smmI
smmI = smmI * (xmax - xmin)
plot(x,f(x))
plot(x,y)
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = 100*(Iexact-smmI)/Iexact
print "I_exact =", Iexact
print "I_Sample_Mean_Method =", smmI
print "Error on I =", Ierr, "%"
```
